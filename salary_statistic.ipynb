```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
salary = pd.read_csv("salary.csv")
#salary.info()
salary.head()
```

```python
salary.describe()
```

```python
#Top 10
salary.nlargest(10, 'Salary')
```

```python
mean = salary.groupby('Race')['Salary'].mean().reset_index()
mean.sort_values(by=['Salary'], inplace=True)
print(mean)

plt.bar(range(len(mean)), mean['Salary'], color="green",  align='center')

plt.xticks(range(len(mean)), mean['Race'], rotation='vertical')
plt.ylabel('Mean Salary')
```

```python
mean = salary.groupby('Gender')['Salary'].mean().reset_index()
mean.sort_values(by=['Salary'], inplace=True)
print(mean)

plt.bar(range(len(mean)), mean['Salary'], color="pink",  align='center')

plt.xticks(range(len(mean)), mean['Gender'], rotation='vertical')
plt.ylabel('Mean Salary')
```
